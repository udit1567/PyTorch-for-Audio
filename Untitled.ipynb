```python
import librosa
import numpy as np

audio, sr = librosa.load("./wav/drums.wav", sr=16000)

mfcc = librosa.feature.mfcc(
    y=audio,
    sr=sr,
    n_mfcc=13
)

print(mfcc.shape)
```
